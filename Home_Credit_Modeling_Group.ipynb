## <a name = "Introduction"> 1. Introduction </a>

Home Credit is an international consumer finance provider primarily focused on lending to individuals with limited or no credit history. They acknowledge that people with little or no credit history often fall prey to untrustworthy lenders. As a result, Home Credit is committed to expanding its financial services to this underserved population to offer a positive and secure borrowing experience. Home Credit makes use of a variety of alternative data, including telco and transactional information, to predict their clients' repayment abilities. Home Credit aims to enhance financial inclusion for the unbanked demographic by offering a secure and positive borrowing experience. To maximize the utility of their data, we will utilize diverse statistical and machine-learning methods for prediction. The objective is to ensure that individuals with the capacity to repay are not denied loans and that loans are structured with a principal, maturity, and repayment schedule that empowers clients to achieve financial success.

The first step is to check the data provided and then do the Exploratory Data Analysis(EDA) to understand the data and its complexities. Then, we will move forward to clean the data so that the data will be ready for applying the models. There are several steps for cleaning data. One step is properly treating the Null values present, like using the imputation method with ‘median’ and ‘mode’ according to the data types. Another step is encoding the categorical columns, like one hot encoding and Label encoding. Another step is to correct the values like outliers and mistaken values. By completing all these steps, we can ensure the data is cleaned and ready to be fitted in Prediction models. Then, moving forward to preparing data for the model, we will join the datasets by properly checking the columns and then will correct the majority class error by using the SMOTE analysis. 

Finally, after fitting the training dataset into the models, we will check the precision, recall, and accuracy values for each model. However, the main performance metric we will use to select the best model is the AUC-ROC score. Along with the AUC-ROC score, we will check the time taken by each model so that we can select the model that executes in less time and a good AUC ROC score.

## <a name = "Dropping_Columns"> 6.2 Dropping Columns </a>

We'll eliminate columns with over 50% missing values from our dataset, but we'll retain 'OWN_CAR_AGE' and 'EXT_SOURCE' columns due to their significant correlation with the target variable.

```python
threshold = 50  # Setting the threshold percentage
# Calculating the total and percentage of missing values for each column
total = application_train.isnull().sum()
percentage = (application_train.isnull().sum() / application_train.isnull().count()) * 100
# Creating a mask to identify columns with more than the threshold percentage of missing values
mask = percentage > threshold
# Selecting the columns to remove
cols_to_remove = total[mask].index
# Identify columns to keep (OWN_CAR_AGE and EXT_SOURCE_1)
cols_to_keep = ['OWN_CAR_AGE', 'EXT_SOURCE_1']
# Remove the specified columns from cols_to_remove
cols_to_remove = [col for col in cols_to_remove if col not in cols_to_keep]
# Remove the columns from the DataFrame
application_train = application_train.drop(columns=cols_to_remove)
```

## <a name = "Joining_Datasets"> 6.5. Joining Datasets </a>

Let's join credit bureau dataset with present application train dataset.

```python
# Define a dictionary with aggregation functions for specific columns in a DataFrame
bureau_agg_dict = {
  'SK_ID_BUREAU': ['count'],
  'DAYS_CREDIT': ['min', 'max', 'mean'],
  'CREDIT_DAY_OVERDUE': ['min', 'max', 'mean'],
  'DAYS_CREDIT_ENDDATE': ['min', 'max', 'mean'],
  'DAYS_ENDDATE_FACT': ['min', 'max', 'mean'],
  'AMT_CREDIT_MAX_OVERDUE': ['max', 'mean'],
  'AMT_CREDIT_SUM': ['max', 'mean', 'sum'],
  'AMT_CREDIT_SUM_DEBT': ['max', 'mean', 'sum'],
  'AMT_CREDIT_SUM_LIMIT': ['max', 'mean', 'sum'],  
  'AMT_CREDIT_SUM_OVERDUE': ['max', 'mean', 'sum'],
  'AMT_ANNUITY': ['max', 'mean', 'sum'],
}
```

```python
# Group the 'bureau_data' DataFrame by the 'SK_ID_CURR' column and apply aggregation functions from 'bureau_agg_dict'
bureau_agg = bureau_data.groupby(['SK_ID_CURR']).agg(bureau_agg_dict)

# Rename the columns in 'bureau_agg' by joining the aggregation function names with column names
bureau_agg.columns = ['BUREAU_'+('_').join(column).upper() for column in bureau_agg.columns.ravel()]

# Reset the index of the 'bureau_agg' DataFrame to make 'SK_ID_CURR' a regular column
bureau_agg = bureau_agg.reset_index()

# Display the first few rows of the resulting 'bureau_agg' DataFrame
bureau_agg.head()
```

```python
# Merge 'application_train' with 'bureau_agg' using a left join on the 'SK_ID_CURR' column
train_bureau_data = application_train.merge(bureau_agg, on='SK_ID_CURR', how='left')

# Merge 'application_test' with 'bureau_agg' using a left join on the 'SK_ID_CURR' column
test_bureau_data = application_test.merge(bureau_agg, on='SK_ID_CURR', how='left')

# Display the first few rows of the 'train_bureau_data' DataFrame
train_bureau_data.head()
```

## <a name = "Impute_Null_Values"> 6.7 Impute Null Values</a>

One common approach to deal with Null Values is imputation, which involves replacing missing values with estimated or calculated values.

Let's first identify categorical columns in both datasets and fill missing values in these columns with the most frequent value (mode) found within each respective column.

For numerical columns, let's replace missing values with the median value of the column.

```python

# Filter categorical columns using select_dtypes
categorical_cols = train_bureau_data.select_dtypes(include='object').columns

# In categorical columns, replacing nulls with the most frequent value of that column
for cat in categorical_cols:
    mode_value = train_bureau_data[cat].mode().values[0]
    train_bureau_data.loc[:, cat].fillna(mode_value, inplace=True)

# In numerical columns, replacing nulls with the median value of that column
for num in train_bureau_data.describe().columns:
    median_value = train_bureau_data[num].median()
    train_bureau_data.loc[:, num].fillna(median_value, inplace=True)
```

## <a name = "Outliers"> 6.8. Outliers </a>

Identifying and addressing outliers is essential to prevent them from skewing statistical results or negatively affecting predictive models. In this section, we will delve into the detection of Outliers.

Winsorization is a statistical technique used to handle outliers in a dataset by capping or limiting extreme values. The process involves replacing values that are beyond certain predefined thresholds (called "limits") with values that are closer to the bulk of the data

```python
# Import the winsorize function
from scipy.stats.mstats import winsorize

# Exclude 'SK_ID_CURR' and 'TARGET' columns from winsorization
numeric_columns = train_bureau_data.select_dtypes(include=[np.number]).drop(columns=['SK_ID_CURR', 'TARGET'])

for column in numeric_columns.columns:
    train_bureau_data[column] = winsorize(train_bureau_data[column], limits=[0.05, 0.05])

# Add back 'SK_ID_CURR' and 'TARGET' columns to the DataFrame
train_bureau_data['SK_ID_CURR'] = train_bureau_data['SK_ID_CURR']
train_bureau_data['TARGET'] = train_bureau_data['TARGET']

print('The new shape is:', train_bureau_data.shape)
```

SMOTE (Synthetic Minority Over-sampling Technique) is a data preprocessing method for addressing imbalanced datasets in machine learning. It works by generating synthetic examples for the minority class by interpolating between existing minority class examples and their nearest neighbors. This helps balance the class distribution and improve the performance of classification models, reducing the risk of bias towards the majority class.

Let's implement Light Gradient Boosting (LightGBM), a predictive modeling technique that boosts accuracy by aggregating predictions from multiple decision trees, akin to Gradient Boosting. Renowned for its efficient implementation, LightGBM is particularly well-suited for large and intricate datasets. The term "gradient" in LightGBM signifies the use of gradient-based optimization methods, employing gradient descent to iteratively adjust model parameters and minimize a specified loss function. The "boosting" process involves sequential training, where each new tree focuses on rectifying errors made by its predecessors. LightGBM employs a histogram-based approach for tree construction, ensuring expedited training times compared to traditional gradient boosting methods. Given our complex home default data, encompassing diverse components like bureau data and credit data, LightGBM proves adept at handling intricate datasets. To address the time-consuming nature of gradient boosting models, we employed GridSearchCV for hyperparameter tuning.


```python
# Import the LightGBM classifier and the LightGBM library
from lightgbm import LGBMClassifier
import lightgbm as lgb
# Replace spaces with underscores in feature names
train_set_resampled.columns = train_set_resampled.columns.str.replace(' ', '_')

# Alternatively, you can remove special characters from feature names using regular expressions
train_set_resampled.columns = train_set_resampled.columns.str.replace('[^a-zA-Z0-9_]', '', regex=True)

# Create an instance of the LightGBM classifier
lgb_mod = LGBMClassifier(num_leaves = 30, random_state=45)

# Start the timer
start_time = time.time()

# Fit the LightGBM model on the resampled training data
lgb_mod.fit(train_set_resampled, train_set_labels_resampled)
```

```python
# predict the test set

lgb_pred = lgb_mod.predict(test_set1)
```

```python
# Import necessary libraries for evaluation
from sklearn import metrics
from sklearn.metrics import accuracy_score, roc_auc_score

# Calculate and print the accuracy score for the model's predictions on the test set

predictions=lgb_mod.predict(train_set_resampled)
# accuracy
accuracy = accuracy_score(test_set_labels1, lgb_pred)
print("Train accuracy:",accuracy_score(train_set_labels_resampled,predictions))
print("Test Accuracy:", accuracy)
# Calculate and print the AUC-ROC score (area under the receiver operating characteristic curve)
auc_roc = roc_auc_score(test_set_labels1, lgb_mod.predict_proba(test_set1)[:, 1])
print("AUC-ROC:", auc_roc)
# Calculate and print the runtime
end_time = time.time()
runtime = end_time - start_time
print("Runtime:", runtime, "seconds")
```

```python
# Confusion Matrix
from sklearn.metrics import  confusion_matrix
cm_lgb = confusion_matrix(test_set_labels1, lgb_pred)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(cm_lgb)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm_lgb[i, j], ha='center', va='center', color='red')

# Classification Report
from sklearn.metrics import  classification_report
cr_lgb = classification_report(test_set_labels1, lgb_pred)
```

The precision score indicates the rate of false positives. It is calculated as the ratio of true positives to the sum of true positives and false positives. High precision indicates a low rate of false positives. The recall score indicates the rate of false negatives. It is calculated as the ratio of true positives to the sum of true positives and false negatives. High recall indicates a low rate of false negatives. Whereas the F1 score is the harmonic mean of precision and recall. Accuracy is the overall accuracy of the model. We will consider the weighted average of precision and recall as it takes all class values by each class weight. Both Precision and recall have to be high for the most accurate results.
